```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

driver = webdriver.Chrome()

Reviews = {
    "Name": [],
    "Review": [],
    "OverAll": [],
    "Food": [],
    "Service": [],
    "Ambience": [],
    "Date": []  
}

ReviewsReq = 2000

def DataScrapping(base_url):
    currentPage = 1
    reviews_scraped = 0  

    while reviews_scraped < ReviewsReq:
        page_url = f"{base_url}?page={currentPage}"
        driver.get(page_url)
        
        try:
            waitTime = WebDriverWait(driver, 5)
            waitTime.until(EC.presence_of_element_located((By.XPATH, "//*[@id='restProfileReviewsContent']")))

            for i in range(1, 11): 
                if reviews_scraped >= ReviewsReq:
                    break 
                
                try:
                    name = driver.find_element(By.XPATH, f'//*[@id="restProfileReviewsContent"]/li[{i}]/section/p[1]').text
                    review = driver.find_element(By.XPATH, f'//*[@id="restProfileReviewsContent"]/li[{i}]/div/div[2]/span[1]').text
                    overall = driver.find_element(By.XPATH, f'//*[@id="restProfileReviewsContent"]/li[{i}]/div/ol/li[1]/span').text
                    food = driver.find_element(By.XPATH, f'//*[@id="restProfileReviewsContent"]/li[{i}]/div/ol/li[2]/span').text
                    service = driver.find_element(By.XPATH, f'//*[@id="restProfileReviewsContent"]/li[{i}]/div/ol/li[3]/span').text
                    ambience = driver.find_element(By.XPATH, f'//*[@id="restProfileReviewsContent"]/li[{i}]/div/ol/li[4]/span').text
                    date = driver.find_element(By.XPATH, f'//*[@id="restProfileReviewsContent"]/li[{i}]/div/div[1]/p').text 
                    
                    Reviews['Name'].append(name)
                    Reviews['Review'].append(review)
                    Reviews['OverAll'].append(overall)
                    Reviews['Food'].append(food)
                    Reviews['Service'].append(service)
                    Reviews['Ambience'].append(ambience)
                    Reviews['Date'].append(date) 

                    reviews_scraped += 1

                except Exception as e:
                    print(f"Error scraping review {i} on page {currentPage}: {e}")
                    continue 

        except Exception as e:
            print(f"Error loading page {currentPage}: {e}")

        WebDriverWait(driver, 10)
        currentPage += 1

DataScrapping("https://www.opentable.com/r/magic-hour-rooftop-bar-and-lounge-new-york")

df = pd.DataFrame(Reviews)
print(df)


```

```python
def ModelWorking(Reviews):

    client = anthropic.Anthropic()
    message = client.messages.create(
        model="claude-3-5-sonnet-20241022",
        max_tokens=1000,
        temperature=0,

    system = f"""
    You are an expert review analyst.
    Analyze the following review and extract the following details clearly:
    1. Comments specifically about the food quality (taste, presentation, variety, freshness, etc.).
    2. Comments specifically about the staff/service (attitude, responsiveness, professionalism, etc.).
    3. For each category, determine if the feedback is positive, negative, or mixed. 
    4. Provide an overall sentiment analysis of the review, classifying it as positive, negative, or Neutral.
    Focus strictly on the review content and avoid including irrelevant data or personal information. Try not to add hallucinations at in any review analysis
    Review: "{Reviews}"
""",
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "text", 
                       
                        "text": Reviews
                    }
                ]
            }
        ]
    )
    return message.content
    
```

```python
import pandas as pd  

review_data = []
processed_count = 0  

for i in range(910):
    if i <= 850:
        continue

    try:
        review = data_frame.iloc[i]["Review"]
        output = ModelWorking(review)  

        text = output[0].text  
        lines = text.split("\n")

       
        food_quality = []
        staff_service = []
        analysis = []

        section = None
        for line in lines:
            if "Food Quality" in line:
                section = "Food Quality"
            elif "Staff/Service" in line or "Service" in line:
                section = "Staff & Service"
            elif "Overall" in line or "Analysis" in line:
                section = "Analysis"
            elif line.strip():  # Skip empty lines
                if section == "Food Quality":
                    food_quality.append(line.strip())
                elif section == "Staff & Service":
                    staff_service.append(line.strip())
                elif section == "Analysis":
                    analysis.append(line.strip())

        review_data.append({
            "Food Quality": " ".join(food_quality),
            "Staff & Service": " ".join(staff_service),
            "Analysis": " ".join(analysis)
        })

        processed_count += 1  # Increment counter
        print(f"Processed {i}/{100} reviews successfully.")

    except Exception as e:
        # Catch any exception, log the error, and skip to the next review
        print(f"Error processing review {i}: {str(e)}. Skipping...")


parsed_reviews_df = pd.DataFrame(review_data)
print(f"Processing complete. Total reviews processed: {processed_count}")
parsed_reviews_df.to_csv("parsed_reviews.csv", index=False)
```
